```python
#rename columns
df = df.rename(columns={"InitialDateReported": "IDR", "MostRecentDateReported": "MRDR","BrandName":"Brand" }) 
df.head()
```

```python
df.isnull().sum() #show the number of rows containing null value
```

```python
df = df.dropna() #drop null values
df.isnull().sum()
```

```python
df.describe() #statistical summary of the data
```

```python
df['ChemicalCount'].value_counts() #show the count of unique values
```

```python
df['CompanyId'].value_counts() #show the count of unique values
```

```python
df['ChemicalId'].value_counts() #show the count of unique values
```

```python
df = df[['CompanyName','CompanyId','Brand','PrimaryCategory','ProductName','ChemicalCount']] # keep specific columns 
df.head()
```

```python
df.set_index('CompanyId',inplace=True) #define the index
df.head()
```

```python
df.to_csv('clean_data.csv') #store the cleaned data to local computer
```

```python
sum_cases_by_date = df.groupby('CompanyId')['ChemicalCount'].sum() #aggregate the total case number per day
sum_cases_by_date.head()
```

```python
sum_cases_by_date.plot() #use a line chart to show the result
```
